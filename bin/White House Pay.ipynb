```python
df['titles'] = df['Position Title'].str.split().apply(lambda x: [item for item in x if item.lower() not in stop])
df['titles'] = df['titles'].apply(lambda x: " ".join(x))
```

```python
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(analyzer = 'word',
                            ngram_range = (2,2),
                            max_features = 5000)
job_feats = vectorizer.fit_transform(df['titles'])
```
