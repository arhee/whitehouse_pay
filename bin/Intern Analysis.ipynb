```python
import pandas as pd
import ggplot
import re

df = pd.read_csv('interns.csv')
```

```python
def get_school(string):
    schooltxt = string.split(';')[-1]
    return schooltxt.rsplit(',',1)
```

```python
def get_home(string):
    mat = re.search(':.*;', string)
    city, state = map(str.strip, mat.group()[2:-1].split(','))[:]
    return city, state
```

```python
def clean_colleges(college):
    collegeList = ['harvard', 'duke', 'johns hopkins', 'princeton', 'rutgers', 'michigan state', 'georgetown', 'george washington',\
                  'yale', 'stanford', 'american', 'emory', 'washburn', 'texas southern']
    
    shortForm = {'new york university':'New York University', 'washington university in': 'Washington University in St. Louis',\
                'texas school of':'University of Texas at Austin', 'berkeley':'University of California, Berkeley', 'boulder':'University of Colorado at Boulder',\
                'ann arbor':'University of Michigan', 'virginia school':'University of Virginia', 'university of wisconsin':'University of Wisconsin, Madison'}

    college = re.sub('The', '', college)
    for item in collegeList:
        if item in college.lower():
            return item.title() + ' University'
        
    for item, longname in shortForm.iteritems():
        if item in college.lower():
            return longname
    return college.strip()
```

```python
def check_alphnum(char):
    char = ord(char)
    if (char >= 48) & (char <= 57):
        return True
    elif (char >= 65) & (char <= 90):
        return True
    elif (char >= 97) & (char <= 122):
        return True
    elif char == 32:
        return True
    else:
        return False
```

```python


names = []
raw_data = df['Raw Data']
for name in raw_data:
    full_name = name.split('Hometown:')[0].strip().split(',')
    full_name = " ".join(full_name[::-1])
    full_name = full_name.lower()
    full_name = "".join([char for char in full_name if check_alphnum(char)])
    names.append(full_name.strip())
```

```python
names
```

```python
df['Cleaned_Names'] = names
```

```python
pd.set_option('display.height', 500)
pd.set_option('display.max_rows', 500)
```

```python
df['Home_City'], df['Home_State'] = zip(*df['Raw Data'].map(get_home))
```

```python
df['Raw_College'], df['College_State'] = zip(*df['Raw Data'].map(get_school))
```

```python
df['College'] = df.Raw_College.apply(clean_colleges)
```

```python
df.rename(columns={'Cleaned_Names':'Cleaned_Name'})
```

```python
df.to_csv('Interns_combined.csv')
```
